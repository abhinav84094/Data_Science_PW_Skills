```python
#stacked histogram >> two histogram together
data1 = np.random.normal(0,1,5000)
data2 = np.random.normal(3,2,5000)
data1,data2
```

```python
plt.hist([data1,data2], bins=20, color=['orange', 'green'], label=['data1', 'data2'])
plt.xlabel("Values")
plt.ylabel("Frequency")
plt.title("Stacked Histogram")
plt.show()
```

```python
# pie chart >> used to understand percent distribution of data
langs = ["C", "Python", "Java"]
students = [20,100,40]

plt.pie(students, labels=langs)
plt.title("Pie Chart")
plt.show()
```

```python
plt.pie(students, labels=langs, autopct="%1.1f%%")
plt.title("Pie Chart")
plt.show()
```

```python
explode = (0.0, 0.1, 0.1)
langs = ["C", "Python", "Java"]
students = [20,100,40]

plt.pie(students, labels=langs , explode=explode)
plt.title("Pie Chart")
plt.show()
```

```python
explode = (0.0, 0.1, 0.1)
langs = ["C", "Python", "Java"]
students = [20,100,40]

plt.pie(students, labels=langs , explode=explode, shadow=True)
plt.title("Pie Chart")
plt.show()
```
