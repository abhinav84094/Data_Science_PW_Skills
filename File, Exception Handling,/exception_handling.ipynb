```python
# custom exception class

class ValidateSalary(Exception):   # Exception is the parent class where all type of exception is 
    def __init__(self, msg):
        self.msg = msg
```

```python
#custom exception

def validate_salary(salary):
    if salary <= 0:
        raise ValidateSalary("negative Salary is not Possible")
    elif salary >= 30000000:
        raise ValidateSalary("Salary is Unexpected")
    else :
        print("Salary is valid")


validate_salary(5000)
```

```python
validate_salary(500000000)
```

```python
try:
    validate_salary(0)
except ValidateSalary as e:
    print(e)
```

```python
try:
    validate_salary(-44)
except ValidateSalary as e:
    print(e)
```

```python
try:
    validate_salary(400000000)
except ValidateSalary as e:
    print(e)
```
