```python
# data interpolation >> it is a process of istimating unknown values from the known values

# Linear Interpolation
# Cubic Interpolation
# Polynomial Interpolation
```

```python
import numpy as np
import matplotlib.pyplot as plt 

x = np.array([1,2,3,4,5])
y = np.array([1,3,5,7,9])

plt.scatter(x, y)
```

```python
# Linear Interpolation >> Linear in nature

x_new  = np.linspace(1,5, 10) # created new x_values

y_interp = np.interp(x_new, x, y)

y_interp, x_new
```

```python
plt.scatter(x_new, y_interp)
```

```python
# cubic interpolation

x = np.array([1,2,3,4,5])
y = np.array([1,8,27,64,125])

plt.scatter(x, y)
```

```python
from scipy.interpolate import interp1d

f = interp1d(x, y, kind='cubic')

f
```

```python
x_new = np.linspace(1,5,10)
y_interp = f(x_new)

y_interp, x_new
```

```python
plt.scatter(x_new, y_interp)
```

```python
# polynomial interpolation

x = np.array([1,2,3,4,5])
y = np.array([1,4,9,1,25])

p = np.polyfit(x,y, 2)

x_new = np.linspace(x,y ,10)

y_interp = np.polyval(p, x_new)

plt.scatter(x,y)
```

```python
plt.scatter(x_new, y_interp)
```
