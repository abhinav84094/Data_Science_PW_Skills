```python
# how to debug >> breaking the code(checkpoint), executing the variables using print statement

def divide(a, b):
    print(a, "divide by ", b)
    return a/b

for i in range(-10, 10):
    print(divide(i, i+1))
```

```python
def divide(a, b):
    print(a, "divide by ", b)
    if b==0:
        return "division by 0, it will throw zeroDivisionEror"
    return a/b


for i in range(-10, 10):
    print(divide(i, i+1))
```

```python
# print statement is not a pythonic way to debug
# print statement is good for simple scripts but not for complex
# in console/terminal all the print statements will be lost
# use logging instead of print statement

#logging >> it records the state and flow of your program/code/software
# it is useful for understanding , monitoring and debugging of your code
# it shows how program behaves over time

# Analogy >> diary entry since childhood to now >> will understand how evolved
# similarly in complex scripts to understand how code is changing the result over time , you can log the specific steps
```

```python
# logging supports different levels of logging which helps to categorize messages based on their severity
import logging

logging.basicConfig(filename='test.log', level=logging.INFO)
```

```python
logging.info("This is normal information about the software")
```

```python
logging.warning("This can be empty list here")
```

```python
logging.debug("The length of list is ")
```

```python
logging.critical("The software has stopped running")
```

```python
logging.error("Some error has happened")
```

```python
logging.shutdown()
```

```python
# DEBUG >> Lowest level of logging , give detailed information of any variable
# INFO >> This is used to convey that the code is working as expected
# WARNING >> Used to indicate that something unexpected happened or potential isuue in the code
# ERROR >> Serious problem
# CRITICAL >> Highest level, termination of program, code, software

# above level of logging is executed in the same order >> debug being lowest priority and critical being highest priority
```

```python
import logging

logging.basicConfig(filename='test_low_level.log', level=logging.DEBUG, format = '%(asctime)s %(message)s')
```

```python
logging.debug("The length of list is ")
logging.info("This is normal information about the software")
logging.warning("This can be empty list here")
logging.error("Some error has happened")
logging.critical("The software has stopped running")
```

```python
import logging 

logging.basicConfig(filename='program.log', level=logging.DEBUG, format = '%(asctime)s %(levelname)s %(message)s')
```

```python
l = [1,"hello",[2, "world"], 4, [3, "Python"]]

l1_int = []
l2_str = []

for i in l:
    logging.info(f"processing each element {i}")

    if type(i)==list:
        for j in i:
            logging.info(f"processing sublist element: {j}")
            if type(j) == int:
                l1_int.append(j)
            elif type(j) == str:
                l2_str.append(j)

    elif type(i)==int:
        l1_int.append(i)
    else :
        l2_str.append(i)
    
logging.info(f"the result is {l1_int} , {l2_str}")
logging.shutdown()
```
