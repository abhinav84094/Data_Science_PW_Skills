```python
start = pd.Timestamp.now()
for i in range(100000000):
    pass

end = pd.Timestamp.now()

print("Time taken to execution this loop = ", end-start)
```

```python
#another way of time
import time
start = time.time()
for i in range(100000000):
    pass
end = time.time()
print("Time taken to execution of this loop = ", end-start)
```

```python
#another way of time
import time
start = time.perf_counter()
for i in range(100000000):
    pass
end = time.perf_counter()
print("Time taken to execution of this loop = ", end-start)
```
